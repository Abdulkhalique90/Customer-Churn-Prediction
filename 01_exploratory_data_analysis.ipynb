# Customer Churn Prediction - Exploratory Data Analysis (EDA)

## Overview
This notebook performs comprehensive exploratory data analysis on the Telco Customer Churn dataset.

### Goals:
1. Load and explore the dataset
2. Understand data types and missing values
3. Analyze churn distribution
4. Perform univariate analysis
5. Perform bivariate analysis
6. Analyze feature correlations
7. Extract key insights

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')

# Set styles
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print('✅ Libraries imported successfully')
```

## 2. Load Dataset

```python
# Load data
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

print(f'✅ Dataset loaded successfully!')
print(f'Shape: {df.shape}')
print(f'Rows: {df.shape[0]}')
print(f'Columns: {df.shape[1]}')
```

## 3. Dataset Overview

```python
# Display first rows
print('First 5 rows:')
df.head()
```

```python
# Data info
print('\nData Types and Info:')
df.info()
```

```python
# Statistical summary
print('\nStatistical Summary:')
df.describe()
```

## 4. Check Missing and Duplicate Values

```python
# Missing values
missing = df.isnull().sum()
print('Missing Values:')
print(missing[missing > 0] if missing.sum() > 0 else 'No missing values!')

# Duplicates
print(f'\nDuplicate Rows: {df.duplicated().sum()}')
```

## 5. Target Variable Analysis - Churn Distribution

```python
# Churn distribution
churn_counts = df['Churn'].value_counts()
print('Churn Distribution:')
print(churn_counts)
print(f'\nChurn Rate: {(churn_counts["Yes"] / len(df) * 100):.2f}%')

# Visualization
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
sns.countplot(data=df, x='Churn', ax=ax[0], palette=['#1f77b4', '#ff7f0e'])
ax[0].set_title('Churn Distribution', fontsize=14, fontweight='bold')
ax[0].set_ylabel('Count')

# Pie chart
ax[1].pie(churn_counts.values, labels=churn_counts.index, autopct='%1.1f%%',
         colors=['#1f77b4', '#ff7f0e'], startangle=90)
ax[1].set_title('Churn Percentage', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print('✅ Churn analysis complete')
```

## 6. Univariate Analysis - Numerical Features

```python
# Get numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f'Numerical Columns: {numerical_cols}')

# Distribution plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for idx, col in enumerate(numerical_cols[:4]):
    ax = axes[idx // 2, idx % 2]
    ax.hist(df[col], bins=30, color='#1f77b4', edgecolor='black', alpha=0.7)
    ax.set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print('✅ Univariate analysis complete')
```

## 7. Univariate Analysis - Categorical Features

```python
# Get categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f'Categorical Columns: {categorical_cols}')

# Display value counts for each categorical column
for col in categorical_cols[:3]:
    print(f'\n{col}:')
    print(df[col].value_counts())
```

## 8. Bivariate Analysis - Churn vs Key Features

```python
# Churn vs Contract Type
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Contract vs Churn
sns.countplot(data=df, x='Contract', hue='Churn', ax=axes[0, 0],
             palette=['#1f77b4', '#ff7f0e'])
axes[0, 0].set_title('Churn by Contract Type', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Count')

# Internet Service vs Churn
sns.countplot(data=df, x='InternetService', hue='Churn', ax=axes[0, 1],
             palette=['#1f77b4', '#ff7f0e'])
axes[0, 1].set_title('Churn by Internet Service', fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel('Count')

# Payment Method vs Churn
sns.countplot(data=df, x='PaymentMethod', hue='Churn', ax=axes[1, 0],
             palette=['#1f77b4', '#ff7f0e'])
axes[1, 0].set_title('Churn by Payment Method', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Count')
axes[1, 0].tick_params(axis='x', rotation=45)

# Tenure vs Churn
sns.boxplot(data=df, x='Churn', y='tenure', ax=axes[1, 1],
           palette=['#1f77b4', '#ff7f0e'])
axes[1, 1].set_title('Tenure by Churn Status', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Tenure (months)')

plt.tight_layout()
plt.show()

print('✅ Bivariate analysis complete')
```

## 9. Churn Rate by Features

```python
# Calculate churn rate by Contract
print('Churn Rate by Contract Type:')
contract_churn = df.groupby('Contract')['Churn'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).sort_values(ascending=False)
print(contract_churn)

print('\nChurn Rate by Internet Service:')
internet_churn = df.groupby('InternetService')['Churn'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).sort_values(ascending=False)
print(internet_churn)

print('\nChurn Rate by Payment Method:')
payment_churn = df.groupby('PaymentMethod')['Churn'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).sort_values(ascending=False)
print(payment_churn)
```

## 10. Correlation Analysis

```python
# Select numerical columns
numerical_df = df.select_dtypes(include=[np.number])

# Correlation matrix
corr_matrix = numerical_df.corr()
print('Correlation Matrix:')
print(corr_matrix)

# Heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
           square=True, ax=ax, cbar_kws={'label': 'Correlation'})
ax.set_title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('✅ Correlation analysis complete')
```

## 11. Key Insights Summary

```python
print('\n' + '='*70)
print('🎯 KEY INSIGHTS FROM EDA')
print('='*70)

print('\n1. CHURN OVERVIEW:')
churn_rate = (churn_counts['Yes'] / len(df) * 100)
print(f'   - Overall Churn Rate: {churn_rate:.2f}%')
print(f'   - Customers Staying: {churn_counts["No"]:,}')
print(f'   - Customers Leaving: {churn_counts["Yes"]:,}')

print('\n2. HIGH CHURN RISK SEGMENTS:')
print(f'   - Fiber Optic Users: {internet_churn["Fiber optic"]:.1f}% churn')
print(f'   - Month-to-Month Contracts: {contract_churn["Month-to-month"]:.1f}% churn')
print(f'   - Electronic Check Payment: {payment_churn["Electronic check"]:.1f}% churn')

print('\n3. TENURE IMPACT:')
avg_tenure_churn = df[df['Churn'] == 'Yes']['tenure'].mean()
avg_tenure_stay = df[df['Churn'] == 'No']['tenure'].mean()
print(f'   - Avg Tenure (Churned): {avg_tenure_churn:.1f} months')
print(f'   - Avg Tenure (Stayed): {avg_tenure_stay:.1f} months')
print(f'   - New customers (< 6 months) have high churn')

print('\n4. RETENTION OPPORTUNITIES:')
print('   ✓ Convert Month-to-Month to longer contracts')
print('   ✓ Improve Fiber Optic service quality')
print('   ✓ Special onboarding for first 6 months')
print('   ✓ Encourage bank transfer payments')

print('\n' + '='*70)
```
